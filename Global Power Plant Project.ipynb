# <div style="background-color:skyblue; padding: 10px; font-family: 'Times New Roman', sans-serif; color: black; text-align: center">___Project Description üåêüí°___</div>

The Global Power Plant Database is a groundbreaking open-source initiative that consolidates information on power plants worldwide. With data covering approximately 35,000 power plants across 167 countries, the database offers a centralized platform for streamlined navigation, comparison, and analysis. It encompasses a diverse range of power sources, including thermal (coal, gas, oil, nuclear, biomass, waste, geothermal) and renewables (hydro, wind, solar). Each entry is geolocated, providing essential details such as plant capacity, generation, ownership, and fuel type. Continuously updated, this comprehensive resource facilitates in-depth insights and analysis.

### Key Attributes of the Database üìä
The database incorporates various indicators, including:

- __country (text):__ 3 character country code (ISO 3166-1 alpha-3)
- __country_long (text):__ Longer form of the country designation
- __name (text):__ Name or title of the power plant (Romanized form)
- __gppd_idnr (text):__ 10 or 12 character identifier for the power plant
- __capacity_mw (number):__ Electrical generating capacity in megawatts
- __latitude (number):__ Geolocation in decimal degrees (WGS84)
- __longitude (number):__ Geolocation in decimal degrees (WGS84)
- __primary_fuel (text):__ Energy source used in primary electricity generation or export
- __other_fuel1,2,3 (text):__ Energy source and Additional energy source used in electricity generation or export
- __commissioning_year (number):__ Year of plant operation, weighted by unit-capacity when data is available
- __owner (text):__ Majority shareholder of the power plant, generally in Romanized form
- __source (text):__ Entity reporting the data; could be an organization, report, or document, generally in Romanized form
- __url (text):__ Web document corresponding to the source field
- __geolocation_source (text):__ Attribution for geolocation information
- __wepp_id (text):__ A reference to a unique plant identifier in the widely-used PLATTS-WEPP database.
- __year_of_capacity_data (number):__ Year the capacity information was reported
- __generation_gwh_2013 to 2019 (number):__ Electricity generation in gigawatt-hours reported for respective years.
- __generation_data_source (text):__ Attribution for the reported generation information
- __estimated_generation_gwh_2013 to 2017 (number):__ Estimated electricity generation in gigawatt-hours for respective years
- __estimated_generation_note_2013 to 2017 (text):__ Label of the model/method used to estimate generation for respective years 


### Fuel Type Aggregation üõ¢Ô∏è
Fuel types in the database are aggregated based on common categories, facilitating easier analysis and classification.

### Predictions üîÆ
1. __Primary Fuel Prediction:__ Using the data, predict the primary fuel type for each power plant.
2. __Capacity Prediction:__ Predict the electrical generating capacity (capacity_mw) for power plants.

```python
# Finding best random state
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
maxAccu = 0
maxRS = 0
for i in range(1,100):
    x_train, x_test, y_train, y_test = train_test_split(x_res,y_res, test_size=.25, random_state=i)
    mod= DecisionTreeClassifier()
    mod.fit(x_train, y_train)
    pred = mod.predict(x_test)
    acc=accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu, "on Random_state ", maxRS)
```

```python
# Finding the best random state
from sklearn.tree import DecisionTreeRegressor
maxAccu = 0
maxRS = 0
for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.30, random_state=i)
    mod= DecisionTreeRegressor()
    mod.fit(x_train, y_train)
    pred = mod.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu, "on Random_state ", maxRS)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scalar, y, test_size=0.2, random_state = 69)
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
# Linear Regression
Lr = LinearRegression()
Lr.fit(x_train, y_train)
pred_test = Lr.predict(x_test)
print("Linear Regression R2 Score:", r2_score(y_test, pred_test))

cv_score_lr = cross_val_score(Lr, x, y, cv=5).mean()
print("Linear Regression CV Score:", cv_score_lr)
```

```python
# Lasso Regression
parameters_lasso = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10], 'random_state': list(range(0, 10))}
ls = Lasso()
clf_lasso = GridSearchCV(ls, parameters_lasso)
clf_lasso.fit(x_train, y_train)
print("Lasso Regression Best Params:", clf_lasso.best_params_)

ls = Lasso(alpha=clf_lasso.best_params_['alpha'], random_state=clf_lasso.best_params_['random_state'])
ls.fit(x_train, y_train)
pred_ls = ls.predict(x_test)
r2_ls = r2_score(y_test, pred_ls)
cv_score_ls = cross_val_score(ls, x, y, cv=5).mean()
print("Lasso Regression R2 Score:", r2_ls)
print("Lasso Regression CV Score:", cv_score_ls)
```

```python
# Decision Tree
dtree = DecisionTreeRegressor()
dtree.fit(x_train, y_train)
pred_dtree = dtree.predict(x_test)
r2_dtree = r2_score(y_test, pred_dtree)
cv_score_dtree = cross_val_score(dtree, x, y, cv=5).mean()
print("Decision Tree R2 Score:", r2_dtree)
print("Decision Tree CV Score:", cv_score_dtree)
```

```python
#Random Forest Regressor
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters ={'criterion':['absolute_error', 'friedman_mse'], 'max_features':["auto","sqrt","log2"]}
Rfr= RandomForestRegressor()
clf =GridSearchCV(Rfr, parameters)
clf.fit(x_train, y_train)

print(clf.best_params_)
```
